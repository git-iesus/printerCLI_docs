```python
# Configurações iniciais

import pandas as pd
import pyautogui as ptg
from time import sleep

df = pd.read_excel('printer.xlsx')
impressao = df['QUANT']
indice = 1

# Definir função

def imprimir(intervalo, folha):
    sleep(0.5)
    ptg.hotkey('ctrl','p')
    sleep(2.0)

    ptg.press('tab')
    sleep(0.1)
    ptg.press('tab')
    sleep(0.1)
    ptg.press('tab')
    sleep(0.1)
    ptg.press('tab')
    sleep(0.1)
    ptg.press('tab')
    sleep(0.1)
    ptg.press('tab')
    sleep(0.1)

    ptg.press('down')
    sleep(0.1)
    ptg.press('down')
    sleep(0.1)
    ptg.press('down')
    sleep(1.0)

    ptg.typewrite(intervalo)
    sleep(0.1)
    ptg.press('tab')
    sleep(0.1)
    ptg.typewrite(folha)
    sleep(0.1)
    ptg.press('enter')
    sleep(0.1)
```

```python
# Abrir o arquivo Merge Docs

ptg.hotkey('win','r')
sleep(0.5)
ptg.typewrite('C:\\Users\\iesus.candido_petfiv\\Desktop\\GitHub\\printerCLI_docs\\merged_docs.pdf')
ptg.press('enter')
sleep(2.5)


#Verificação da planilha

for linha in impressao:
    indice +=1
    if linha == 0:
        # Não há nada para imprimir
        continue
    else:
        match indice:
            case 2:  #CLASSIFICACAO
                imprimir('10',str(linha))
            case 3:  #PRODUTO INTERMEDIÁRIO
                imprimir('11',str(linha))
            case 4:  #MOINHO MARTELO
                imprimir('12',str(linha))
            case 5:  #PÓ
                imprimir('13',str(linha))
            case 6:  #VARREDURA
                imprimir('14',str(linha))
            case 7:  #setup de envase
                imprimir('27',str(linha))
            case 8:  #carregamento de amido
                imprimir('28',str(linha))
            case 9:  #processo formulação
                imprimir('29',str(linha))
            case 10: #abastecimento externo
                imprimir('8',str(linha))
            case 11: #pó do secador
                imprimir('9',str(linha))
            case 12: #processo filtragem
                imprimir('17',str(linha))
            case 13: #processo secagem
                imprimir('18',str(linha))
            case 14: #resfriamento e segregação
                imprimir('19',str(linha))
            case 15: #moega do secador
                imprimir('24',str(linha))
            case 16: #peneiras retrabalho
                imprimir('20-21',str(linha))
            case 17: #envase
                imprimir('22-23',str(linha))
            case 18: #consolid
                imprimir('25-26',str(linha))
            case 19: #vibros 3 e 4
                imprimir('1-3',str(linha))
            case 20: #estrechadeira
                imprimir('4-5',str(linha))
            case 21: #abastecimento interno
                imprimir('6-7',str(linha))
            case 22: #moinho
                imprimir('30-31',str(linha))

ptg.alert('Automação do printerCLI concluída!')
```
